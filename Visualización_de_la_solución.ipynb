# Ejercicio 2 - Visualización de la solución

```python
import numpy as np    # Biblioteca numérica
import matplotlib.pyplot as plt    # Biblioteca para gráficas

def f(x,y):    # Define ecuación diferencial
    return x + y    # Retorna pendiente

x = 0    # Valor inicial
y = 1    # Condición inicial

h = 0.1    # Paso numérico

x_vals = []    # Lista para almacenar valores de x
y_vals = []    # Lista para almacenar valores de y

for i in range(20):    # Ejecuta 20 pasos

    x_vals.append(x)    # Guarda valor actual de x

    y_vals.append(y)    # Guarda valor actual de y

    y = y + h*f(x,y)    # Método de Euler

    x = x + h    # Incrementa x

plt.plot(x_vals,y_vals)    # Grafica solución aproximada

plt.xlabel("x")    # Etiqueta eje x

plt.ylabel("y")    # Etiqueta eje y

plt.title("Solución aproximada de la EDO")    # Título

plt.grid()    # Activa cuadricula

plt.show()    # Muestra gráfica
```

# Ejercicio 1 - Resolver una EDO simple

```python
import numpy as np    # Importa NumPy para cálculos numéricos

def f(x,y):    # Define la ecuación diferencial
    return x + y    # Retorna el valor de dy/dx

x = 0    # Valor inicial de x
y = 1    # Condición inicial y(0)=1

h = 0.1    # Tamaño del paso

for i in range(5):    # Ejecuta 5 iteraciones

    y = y + h*f(x,y)    # Aproximación numérica tipo Euler

    x = x + h    # Avanza al siguiente punto

print("x =",x,"y =",y)    # Muestra resultados
```
